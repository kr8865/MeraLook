```python
def detect_body_type(chest, waist, hip):
    if abs(chest - hip) <= 5 and waist < chest - 10:
        return 'Hourglass'
    elif hip > chest + 5:
        return 'Pear'
    elif waist > chest and waist > hip:
        return 'Apple'
    elif abs(chest - waist) < 6 and abs(hip - waist) < 6:
        return 'Rectangle'
    else:
        return 'Undefined'
```

```python
import pandas as pd
df=pd.read_csv('data.csv')
print(df.head())
```

```python
df['type'] = df.apply(lambda row: detect_body_type(row['Chest'], row['Waist'], row['Hips']), axis=1)
```
